**Descripción de los datos**

Estas son las columnas de nuestro dataset y una descripción de ellas.

- ```transaction_id``` - identificador único de la transacción.
- ```fecha``` - fecha en la que se registró la transacción.
- ```anio``` - año de la transacción.
- ```mes``` - número del mes de la transacción.
- ```nombre_mes``` - nombre del mes de la transacción.
- ```empresa``` - nombre de la empresa asociada a la transacción.
- ```cliente_proveedor``` - cliente o proveedor relacionado con la operación.
- ```categoria``` - categoría principal de la transacción.
- ```subcategoria``` - clasificación más específica dentro de la categoría.
- ```tipo_movimiento``` - tipo de operación financiera, como ingreso o gasto.
- ```importe_limpio``` - importe de la transacción depurado y normalizado.
- ```moneda``` - moneda en la que se realizó la transacción.
- ```canal_pago``` - medio utilizado para realizar el pago.
- ```estado``` - situación actual de la transacción.
- ```centro_coste``` - área o centro de coste asociado a la operación.
- ```descripcion``` - detalle o comentario descriptivo de la transacción.
- ```referencia_externa``` - código o referencia externa vinculada a la operación.
- ```pais``` - país asociado a la transacción.

### Etapa 2: Preparación y transformación de datos

**Conversión de tipos**

````python
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")
df["importe_limpio"] = pd.to_numeric(df["importe_limpio"], errors="coerce")
````

**Métricas principales**

````python
total_transacciones = len(df)
total_importe = df["importe_limpio"].sum()
ingresos = df.loc[df["tipo_movimiento"] == "Ingreso", "importe_limpio"].sum()
gastos = df.loc[df["tipo_movimiento"] == "Gasto", "importe_limpio"].sum()
balance = ingresos - gastos

print(f"""
Total transacciones: {total_transacciones}
Total importe: {total_importe:,.2f}
Ingresos: {ingresos:,.2f}
Gastos: {gastos:,.2f}
Balance: {balance:,.2f}
""")
````

**Estados de las transacciones** [ ESTADO ]

````python
resumen_estado = (
    df.groupby("estado")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .reset_index()
    .sort_values("transacciones", ascending=False)
)

resumen_estado["porcentaje"] = resumen_estado["transacciones"] / resumen_estado["transacciones"].sum() * 100

print(resumen_estado)
````

**Descripción**

Esta sección muestra cuántas transacciones hay en cada estado y cuánto importe representa cada una.
Es el patrón más importante del proyecto, porque permite evaluar cierre operativo, fricción y riesgo financiero.[1]

---

**Empresas con mayor volumen** [ EMPRESAS ]

````python
resumen_empresa = (
    df.groupby("empresa")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .sort_values("importe_total", ascending=False)
    .reset_index()
)

print(resumen_empresa)
````

**Descripción**
  
Aquí se identifican las empresas con mayor actividad y mayor peso financiero. Esto ayuda a priorizar seguimiento y detectar concentraciones relevantes.

---

**Categorías más relevantes** [ CATEGORIA ]

````python
resumen_categoria = (
    df.groupby("categoria")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .sort_values("importe_total", ascending=False)
    .reset_index()
)

print(resumen_categoria)
````

**Descripción**

Las categorías permiten ver en qué tipo de operación se concentra el dinero. Esto aporta contexto para control financiero y análisis de negocio.

---

**Distribución por país** [ PAIS ]

````python
resumen_pais = (
    df.groupby("pais")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .sort_values("importe_total", ascending=False)
    .reset_index()
)

print(resumen_pais)
````

**Descripción**
  
Esta segmentación muestra cómo se reparte la actividad por país. Sirve para detectar concentración geográfica y diferencias operativas.

---

**Distribución por canal de pago** [ CANAL_PAGO ]

````python
resumen_canal = (
    df.groupby("canal_pago")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .sort_values("importe_total", ascending=False)
    .reset_index()
)

print(resumen_canal)
````

**Descripción**

Esta tabla permite identificar los canales de pago más utilizados. También ayuda a explorar si algún canal se asocia con más fricción o menor cierre.

---


**Evolución temporal** [ MES ]

````python
resumen_mes = (
    df.groupby(["anio", "mes", "nombre_mes"])
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .reset_index()
    .sort_values(["anio", "mes"])
)

print(resumen_mes)
````

**Descripción**
  
Esta vista permite estudiar estacionalidad y evolución de la actividad. Es útil para detectar picos de carga y planificación financiera.

### Etapa 4: Clasificación y segmentación de datos

**Tipo de movimiento** [ TIPO_MOVIMIENTO ]

````python
tipo_resumen = (
    df.groupby("tipo_movimiento")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .reset_index()
)

print(tipo_resumen)
````

**Descripción**
  
Separar ingresos y gastos permite entender el comportamiento del flujo financiero. También sirve como base para evaluar balance neto.

---


**Centro de coste** [ CENTRO_COSTE ]

````python
centro_resumen = (
    df.groupby("centro_coste")
    .agg(
        transacciones=("transaction_id", "count"),
        importe_total=("importe_limpio", "sum")
    )
    .reset_index()
    .sort_values("importe_total", ascending=False)
)

print(centro_resumen)
````

**Descripción**
  
Esta segmentación muestra qué áreas internas concentran más actividad o costo. Ayuda a visualizar dónde se generan más transacciones.

---


**Estados por canal** [ CANAL_PAGO ]

````python
estado_canal = (
    df.groupby(["canal_pago", "estado"])
    .agg(transacciones=("transaction_id", "count"))
    .reset_index()
)

print(estado_canal)
````

**Descripción**

Permite ver si algún canal presenta más pendientes, cancelaciones o rechazos. Esto ayuda a detectar puntos de fricción operativa.

---


**Estados por país** [ PAIS ]

````python
estado_pais = (
    df.groupby(["pais", "estado"])
    .agg(transacciones=("transaction_id", "count"))
    .reset_index()
)

print(estado_pais)
````

**Descripción**
  
Sirve para comparar cómo cambia el comportamiento de los estados según el país. Es útil para análisis geográfico de riesgo.

## KPIs útiles

````python
top_importes = (
    df.sort_values("importe_limpio", ascending=False)
    .head(5)[["transaction_id", "empresa", "estado", "importe_limpio", "canal_pago", "pais"]]
)

print(top_importes)
````
